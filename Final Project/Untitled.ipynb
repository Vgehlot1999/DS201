```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
sns.set_style('whitegrid')
import pandas as pd
df = pd.read_csv('SupportedProcessors.csv')
df.head()
```

```python
df = df.drop(['Tag'], axis = 1)
df = df.drop(['Description'], axis = 1)
df = df.drop(['Connected To'], axis = 1)
df = df.drop(['Jack Media'], axis = 1)
df = df.drop(['Jack'], axis = 1)
df = df.drop(['IP'], axis = 1)
df = df.drop(['Client'], axis = 1)
df = df.drop(['NetID'], axis = 1)
df = df.drop(['Fund_S_P'], axis = 1)
df = df.drop(['Empl Num'], axis = 1)
df = df.drop(['Vendor Serial'], axis = 1)
df = df.drop(['MAC_Wired'], axis = 1)
df = df.drop(['MAC_Wireless'], axis = 1)
df = df.drop(['MAC_Other'], axis = 1)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
x = df['Year'].values
y = df['Warranty Year'].values
plot.scatter(x, y)
plt.xlabel('Year Acquired')
plt.ylabel('Warranty Year')
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x.reshape(-1, 1), y)
model_pred = lin_reg.predict(x.reshape(-1,1))
plt.figure(figsize=(10,8));
plt.scatter(x, y);
plt.plot(x, model_pred);
print(r2_score(y, model_pred))
plt.xlabel('Year Acquired')
plt.ylabel('Warranty Year')
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(x.reshape(-1, 1))

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y.reshape(-1, 1))
y_pred = lin_reg_2.predict(X_poly)
plt.figure(figsize=(10,8));
plt.scatter(x, y);
plt.plot(x, y_pred);
print(r2_score(y, y_pred))
```

```python
poly_reg_2 = PolynomialFeatures(degree=3)
X_poly_2 = poly_reg_2.fit_transform(x.reshape(-1, 1))

lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly_2, y.reshape(-1, 1))
y_pred = lin_reg_3.predict(X_poly_2)
plt.figure(figsize=(10,8));
plt.scatter(x, y);
plt.plot(x, y_pred);
print(r2_score(y, y_pred))
```

```python
df6 = df
df6 = df6.drop(['Room'], axis = 1)
df6 = df6.drop(['OS'], axis = 1)
df6 = df6.drop(['Office'], axis = 1)
df6 = df6.drop(['Fund'], axis = 1)
df6 = df6.drop(['Acquired'], axis = 1)
df6 = df6.drop(['Warranty End Date'], axis = 1)
df6 = df6.drop(['Wire/Less'], axis = 1)
df6 = df6.drop(['Building'], axis = 1)
df6.head()
```

```python
x = df6[['Billing', 'Year']].values
y = df['Warranty Year'].values


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 0)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn import metrics
import numpy as np

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)

print('Training Accuracy:')
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

#accuracy of data after having 3 or more neighbors stays identical, so I decided to keep it at 5
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = x_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.8, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1])
plt.title('K-NN (Training set)')
plt.xlabel('Billing')
plt.ylabel('Year')
plt.show()
```

```python
from sklearn.cluster import KMeans

L=list(df6.columns)[2:4]
x=df[L]
x.head()
```

```python
kmeans = KMeans(n_clusters = 7)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
y_kmeans
```

```python
x = np.array(x)
plt.scatter(x[:, 0], x[:, 1],  c = y_kmeans, s = 50, cmap = 'viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'black', s = 200, alpha = 0.5)
plt.xlabel("Year")
plt.ylabel("Warranty Year")
```
